```python
import pygame
import sys
import os

pygame.init()

# Window
WIDTH, HEIGHT = 800, 600
WIN = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Gun Arena")

# Colors
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)
BLUE = (0, 0, 255)
RED = (255, 0, 0)
SILVER = (192, 192, 192)
GOLDEN_RED = (200, 60, 60)
DARK_GREY = (30, 30, 30)

# Fonts
FONT = pygame.font.SysFont("comicsans", 25)
BIG_FONT = pygame.font.SysFont("comicsans", 30)

# Player settings
PLAYER_WIDTH, PLAYER_HEIGHT = 50, 60
VELOCITY = 5
BULLET_VEL = 7
MAX_HEALTH = 100

# Clock
FPS = 60
clock = pygame.time.Clock()

# Load background image
BG_IMAGE = pygame.image.load("C:/Users/Ayush Goyal/Downloads/WhatsApp Image 2025-04-16 at 10.03.04 PM.jpeg")  # Replace with your background image path
BG_IMAGE = pygame.transform.scale(BG_IMAGE, (WIDTH, HEIGHT))

# Load background music
pygame.mixer.music.load("C:/Users/Ayush Goyal/Downloads/doingdamage.mp3")  # Replace with your new background music file
pygame.mixer.music.set_volume(0.3)  # Adjust volume as needed
pygame.mixer.music.play(-1, 0.0)  # Loop indefinitely

# Load player images
PLAYER1_IMG = pygame.image.load("C:/Users/Ayush Goyal/Pictures/WhatsApp Image 2025-04-16 at 9.33.07 PM.jpeg")
PLAYER1_IMG = pygame.transform.scale(PLAYER1_IMG, (PLAYER_WIDTH, PLAYER_HEIGHT))
PLAYER2_IMG = pygame.image.load("C:/Users/Ayush Goyal/Pictures/WhatsApp Image 2025-04-16 at 9.52.02 PM.jpeg")
PLAYER2_IMG = pygame.transform.scale(PLAYER2_IMG, (PLAYER_WIDTH, PLAYER_HEIGHT))

class Fighter:
    def _init_(self, x, y, image, controls):
        self.x = x
        self.y = y
        self.width = PLAYER_WIDTH
        self.height = PLAYER_HEIGHT
        self.image = image
        self.health = MAX_HEALTH
        self.controls = controls
        self.bullets = []

    def draw(self, window):
        window.blit(self.image, (self.x, self.y))
        for bullet in self.bullets:
            pygame.draw.rect(window, (255, 255, 0), bullet)

    def move(self, keys):
        if keys[self.controls["left"]] and self.x - VELOCITY > 0:
            self.x -= VELOCITY
        if keys[self.controls["right"]] and self.x + VELOCITY + self.width < WIDTH:
            self.x += VELOCITY
        if keys[self.controls["up"]] and self.y - VELOCITY > 0:
            self.y -= VELOCITY
        if keys[self.controls["down"]] and self.y + VELOCITY + self.height < HEIGHT:
            self.y += VELOCITY

    def shoot(self):
        bullet = pygame.Rect(self.x + self.width // 2, self.y + self.height // 2, 10, 5)
        self.bullets.append(bullet)

    def handle_bullets(self, other):
        for bullet in self.bullets[:]:
            bullet.x += BULLET_VEL if self.x < other.x else -BULLET_VEL
            if other.x < bullet.x < other.x + other.width and other.y < bullet.y < other.y + other.height:
                other.health -= 10
                self.bullets.remove(bullet)
            elif bullet.x < 0 or bullet.x > WIDTH:
                self.bullets.remove(bullet)

def draw_window(player1, player2):
    WIN.blit(BG_IMAGE, (0, 0))

    # Draw names with golden-reddish background
    names = ["AYUSH", "RISHAANK", "SANSKAR", "SACHDEVA", "ISHVI"]
    for idx, name in enumerate(names):
        name_surface = BIG_FONT.render(name, True, WHITE)
        name_bg = pygame.Surface((name_surface.get_width() + 15, name_surface.get_height() + 6))
        name_bg.fill(GOLDEN_RED)
        y_offset = 20 + idx * (name_surface.get_height() + 8)
        WIN.blit(name_bg, (WIDTH // 2 - name_bg.get_width() // 2, y_offset))
        WIN.blit(name_surface, (WIDTH // 2 - name_surface.get_width() // 2, y_offset + 3))

    # Draw players and bullets
    player1.draw(WIN)
    player2.draw(WIN)

    # Health bars
    pygame.draw.rect(WIN, (220, 20, 60), (10, 10, MAX_HEALTH, 20))
    pygame.draw.rect(WIN, GREEN, (10, 10, player1.health, 20))

    pygame.draw.rect(WIN, (220, 20, 60), (WIDTH - MAX_HEALTH - 10, 10, MAX_HEALTH, 20))
    pygame.draw.rect(WIN, GREEN, (WIDTH - 10 - player2.health, 10, player2.health, 20))

    pygame.display.update()

def display_winner(winner_text):
    text = FONT.render(winner_text, True, WHITE)
    WIN.blit(text, (WIDTH // 2 - text.get_width() // 2, HEIGHT // 2))
    pygame.display.update()
    pygame.time.delay(3000)

def main():
    player1 = Fighter(100, HEIGHT // 2, PLAYER1_IMG, {
        "left": pygame.K_a, "right": pygame.K_d, "up": pygame.K_w, "down": pygame.K_s, "shoot": pygame.K_g
    })
    player2 = Fighter(600, HEIGHT // 2, PLAYER2_IMG, {
        "left": pygame.K_LEFT, "right": pygame.K_RIGHT, "up": pygame.K_UP, "down": pygame.K_DOWN, "shoot": pygame.K_k
    })

    run = True
    while run:
        clock.tick(FPS)
        keys = pygame.key.get_pressed()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
                pygame.quit()
                sys.exit()

            if event.type == pygame.KEYDOWN:
                if event.key == player1.controls["shoot"]:
                    player1.shoot()
                if event.key == player2.controls["shoot"]:
                    player2.shoot()

        player1.move(keys)
        player2.move(keys)

        player1.handle_bullets(player2)
        player2.handle_bullets(player1)

        draw_window(player1, player2)

        if player1.health <= 0:
            display_winner("Player 2 Wins!")
            break
        if player2.health <= 0:
            display_winner("Player 1 Wins!")
            break

    main()

if _name_ == "_main_":
    main()
```
